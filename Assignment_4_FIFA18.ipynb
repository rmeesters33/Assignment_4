De data kolommen waarvoor gekozen is zijn variabelen die een verdediger, middenvelden en aanvaller vaak onderscheiden van elkaar. Dit zijn 7 variabelen, maar het hadden er ook zeker meer kunnen zijn. 

```python
df = df[["Preferred Positions", "Agility", "Dribbling", "Interceptions", "Marking", "Finishing", "Short passing"]]

df.info()
```

```python
from sklearn.preprocessing import normalize

X = df.loc[:, ~df.columns.isin(["Preferred Positions"])] 
y = df["Preferred Positions"] 

X = normalize(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
pd.DataFrame(X_train).head()
```
